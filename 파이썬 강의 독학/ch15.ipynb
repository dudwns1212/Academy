```python
%matplotlib inline
# 한글 깨짐 해결 코드
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('systolic.csv')
sns.lmplot(x='연령', y='BMI', hue='성별', data=df)
plt.show()
```

```python
df2 = df.drop(['환자코드','성별','혈액형'], axis=1)
df2
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df2)
plt.title("생체 데이터 분석")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df2)
plt.title("생체 데이터 분석")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=df2)
plt.title("생체 데이터 분석")
plt.show()
```

```python
g = sns.catplot(data=df, kind='bar', col='BMI', col_wrap=5)
g.fig.set_size_inches(14,10)
g.fig.subplots_adjust(wspace=0.1, hspace=0.1)
plt.show()
```
